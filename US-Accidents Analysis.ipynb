### Start Time

```python
df.columns
```

```python
df.Start_Time
```

```python
df.Start_Time[0]
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
sns.distplot(df.Start_Time.dt.hour , bins=24 , norm_hist=True)

# -- A high percentage of accident occur between 6 am to 10 am(probably people in a hurry to get to work)
# -- And also the number of accidents is high in between 3 pm to 6 pm(probably traffic is high or maybe overspeeding or maybe people going back to home from their work)
# -- But here is the strange thing is that after 6 pm the number of accidents is reducing , because there is always high chances of accidents in night(probably the num most of the people doesnt drive after 6pm)
```

```python
sns.distplot(df.Start_Time.dt.dayofweek , bins=7 , norm_hist=True)
```

Is the distribution of accident by hour the same on weekends as on weekdays.

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour , bins=24 , norm_hist=True)
```

```python
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(mondays_start_time.dt.hour , bins=24 , norm_hist=True)
```

On Sundays, the peak occurs between 10 am & 3 pm , unlike week days

```python
 df.Start_Time.dt.year
```

```python
sns.distplot(sundays_start_time.dt.month , bins=12 , norm_hist=True)
# -- The number of accidents is most happening in between the winter season possibly due to less visibility on roads.
# -- But on the other hand the number of accidents are lower during summer. 
```

```python
df_2019 = df[df.Start_Time.dt.year == 2017]
df_2019_Source1 = df_2019[df_2019.Source == 'Source1']
sns.distplot(df_2019_Source1.Start_Time.dt.month , bins=12 , norm_hist=True)
```
